```python
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
file_path = 'C:/Users/irajn/OneDrive/Desktop/finance/linereg fin.xlsx'  
data = pd.read_excel(file_path)
```

```python
clean_data_google = data.dropna(subset=['Exp ret GOOGL', 'Exp ret NASDAQ'])
X = clean_data_google[['Exp ret NASDAQ']].values
y = clean_data_google['Exp ret GOOGL'].values
```

```python
clean_data_netflix = data.dropna(subset=['Exp ret NFLX', 'Exp ret NASDAQ'])
x_1 = clean_data_netflix[['Exp ret NASDAQ']].values
y_1 = clean_data_netflix['Exp ret NFLX'].values
```

```python
model1 = LinearRegression()
model1.fit(X, y)

# The slope (beta) of the regression line
beta = model1.coef_[0]
print(f"The beta of GOOGL relative to NASDAQ is: {beta}")
```

```python
model2 = LinearRegression()
model2.fit(x_1, y_1)

# The slope (beta) of the regression line
beta = model2.coef_[0]
print(f"The beta of NFLX relative to NASDAQ is: {beta}")
```

```python
# Scatter plot of the data
plt.figure(figsize=(10, 6))  # Adjust the size of your plot
sns.scatterplot(x='Exp ret NASDAQ', y='Exp ret GOOGL', data=clean_data_google, color='blue', label='Data Points')

# Plotting the regression line
# Generate predictions for the line
line_X = np.linspace(X.min(), X.max(), 100)[:, np.newaxis]  # Generating X values
line_y = model1.predict(line_X)  # Predicting Y values based on the model

sns.lineplot(x=line_X[:, 0], y=line_y, color='red', label='Regression Line')

# Additional plot settings
plt.title('Linear Regression: GOOGL vs NASDAQ Returns')
plt.xlabel('NASDAQ Expected Returns')
plt.ylabel('GOOGL Expected Returns')
plt.legend()
plt.show()
```

```python
# Scatter plot of the data
plt.figure(figsize=(10, 6))  # Adjust the size of your plot
sns.scatterplot(x='Exp ret NASDAQ', y='Exp ret GOOGL', data=clean_data_netflix, color='blue', label='Data Points')

# Plotting the regression line
# Generate predictions for the line
line_X = np.linspace(X.min(), X.max(), 100)[:, np.newaxis]  # Generating X values
line_y = model2.predict(line_X)  # Predicting Y values based on the model

sns.lineplot(x=line_X[:, 0], y=line_y, color='red', label='Regression Line')

# Additional plot settings
plt.title('Linear Regression: NFLX vs NASDAQ Returns')
plt.xlabel('NASDAQ Expected Returns')
plt.ylabel('NFLX Expected Returns')
plt.legend()
plt.show()
```
